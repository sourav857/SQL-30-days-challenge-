```python
# 14. How many people have a credit card that expires in 2020?
```

```python
# df.columns'Credit Card'
def ext_year(extract_year):
    parts=extract_year.split('/')
    if len(parts)>1:
        year=parts[1].strip()
        return year
    return none
df['new_year']=df['CC Exp Date'].apply(extract_year)

res=df[df['new_year']=='20']


len(res[res['Credit Card'].notnull()])


```

```python


# 15. What are the top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...)
```

```python
def provide(e_provide):
    parts=e_provide.split('@')
    if len(parts)>1:
        res=parts[1].strip()
        return res
    return none
df['email providers']=df['Email'].apply(provide)
df['email providers'].value_counts().head(5)

```
