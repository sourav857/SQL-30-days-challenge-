```python
# df.head(20)

df['ship_mode'].unique()

df['ship_mode'] = df['ship_mode'].str.replace('Not Available', 'nan')
```

```python
df['discount']=df['discount_percent']*df['list_price']*0.01
```

```python
df['sale_price']=df['list_price']-df['discount']
```

```python
df['profit']=df['sale_price']-df['cost_price']
```
