```python
df1=sns.load_dataset("penguins")
```

```python
df1.head(20)
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1)
plt.show()
```

```python
# show based on sex
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue="sex",
             style="sex",palette="rocket_r",markers=["o",">"])
plt.grid()
plt.title("This is line graph")
plt.show()
```

```python
import pandas as pd
```

```python
df1["island"]
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm,data=df1,hue="sex")
plt.show()
```

```python
order1=["Biscoe","Torgersen","Dream"]
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",order=order1)
plt.show()
```

```python
# hue order ie wan to change order of the order
```

```python
order1=["Biscoe","Torgersen","Dream"]
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",
            order=order1,hue_order=["Female","Male"])
plt.show()
```

```python
order1=["Biscoe","Torgersen","Dream"]
sns.barplot(x="island",y="bill_length_mm",data=df1,color="green")
plt.show()
```
