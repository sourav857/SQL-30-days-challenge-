```python
import pandas as pd
from sqlalchemy import create_engine,text
import mysql.connector as connection

# Establish the MySQL connection
mydb = connection.connect(
    host='localhost', user='root', passwd='GETdata123@720', database='project', use_pure=True)
def extract():
    df_product=pd.read_csv("C:/Program Files/JetBrains/PyCharm 2023.3.4/product.txt")
    df_products_db=pd.read_sql_query('select * from products',con=engine)
    return df_product,df_products_db

def transform(df_product,df_products_db):
    df_merged=pd.merge(df_product,df_products_db,how='left',on='product_id')
    df_insert=df_merged[df_merged['product_name_y'].isna()]
    df_insert_final=df_insert.iloc[:,0:3]
    df_insert_final.columns=df_products_db.columns

    df_update=df_merged[df_merged['product_name_y'].notna()]
    df_update_final=df_update.iloc[:,0:3]
    df_update_final.columns=df_products_db.columns
    return df_insert_final,df_update_final
def load_stage(df_update_final):
    engine = create_engine('mysql+mysqlconnector://root:GETdata123%40720@127.0.0.1:3306/project', echo=True)
    df_update_final.to_sql('products_stage',con=engine,if_exists='replace',index=False)
  
    
def updates():
    engine = create_engine('mysql+mysqlconnector://root:GETdata123%40720@127.0.0.1:3306/project', echo=True)
    query = text('UPDATE products JOIN products_stage ON products.product_id = products_stage.product_id SET products.price = products_stage.price')
    with engine.connect() as con:
        con.execute(query)
        con.commit()


def inserts(df_insert_final):
    engine = create_engine('mysql+mysqlconnector://root:GETdata123%40720@127.0.0.1:3306/project', echo=True)
    df_insert_final.to_sql('products',con=engine,if_exists='append',index=False)
   


 

```
